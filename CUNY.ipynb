# CUNY journalism skills investigation

```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as BS
from random import choice
from time import sleep
```

```python
base_url = "http://www.indeed.com/resumes?q=journalist&l=New+York%2C+NY"
# http://www.indeed.com/resumes?q=journalist&l=New+York%2C+NY&co=US&start=50
# http://www.indeed.com/resumes?q=journalist&l=New+York%2C+NY&co=US&start=100

# Text when pages are exhausted:
# The page you are looking for was not found.
```

## Scrape 'journalist' search results for 'New York, NY'

```python
base = "http://www.indeed.com/resumes?q=journalist&l=New+York%2C+NY&co=US&start="

search_results = []
for n in range(0, 1000, 50): # Manually checked there ~ 950 results
    search_results.append(requests.get(base + str(n)).text)
    sleep(choice([0, 1, 2, 3]))
    
```

```python
resume_urls = []
for result in search_results:
    resume_urls.extend(re.findall('''href="/r/(.+?)"''', r))
```

```python
len(resume_urls)
```

## Scrape individual resume pages and back them up

```python
resume_pages = []
```

```python
for resume_url in resume_urls:
    resume_pages.append(requests.get("http://www.indeed.com/r/"+resume_url).text)
    sleep(choice([0, 1, 2, 3]))
    
```

```python
df = pd.DataFrame(resume_pages, columns=["page"])
df.to_csv("ResumePages.csv")
```

## Extract skills from pages (that have them)

```python
df = pd.read_csv("ResumePages.csv")
resume_pages = df["page"]
```

```python
skills = []
```

```python
for resume in resume_pages:
    soup = BS(str(resume), 'html.parser')
    skills_content = soup.findAll("div", {"class": "section-item skills-content"})
    pat_skill = '''<span class="skill-text">(.+?)</span>'''
    try:
        skills.append(re.findall(pat_skill, str(skills_content[0])))
    except:
        pass
```
